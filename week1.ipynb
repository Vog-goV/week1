```python
# Problem 1

scores = [72, 85, 90, 68, 88]
print(scores)
average = sum(scores)/len(scores)
print(f"Average = {average}")
print(f"Highest = {max(scores)}")
print(f"Lowest = {min(scores)}")
```

```python
# Problem 2

study_hours = [1, 2, 3, 4, 5]
exam_scores = [50, 55, 65, 70, 80]
print(f"Study Hours: {study_hours}\nExam Scores: {exam_scores}")
for i in range(len(study_hours)):
  print(f"{study_hours[i]} hours -> {exam_scores[i]} marks")
```

```python
# Problem 3

def pass_fail(sc):
  if sc >= 60:
    return "Pass"
  else:
    return "Fail"


scores = [45, 67, 82, 59]
for i in range(len(scores)):
  print(f"{scores[i]} -> {pass_fail(scores[i])}")
```

```python
# Problem 4
import matplotlib.pyplot as plt

size = [500, 800, 1000, 1200]
prices = [50, 80, 100, 130]

for i in range(len(size)):
  print(f"{size[i]} -> {prices[i]}K$")
plt.plot(size, prices)
plt.xlabel("Size")
plt.ylabel("Price")
plt.title("Trend")
plt.show()
```

```python
# Problem 5

study_hours = [2, 4, 6, 8]
scores = [55, 65, 75, 85]

for i in range(len(scores)-1, 0 , -1):
  avg = scores[i] - scores[i-1]

prdc = scores[-1]+avg
print(prdc)
```

```python
# Problem 6
scores = [35, 58, 62, 74, 89, 45]

low = []
high = []
medium = []

for i in range(len(scores)):
  if scores[i] < 50:
    low.append(scores[i])
  elif scores[i] >= 50 and scores[i] <= 70:
    medium.append(scores[i])
  else:
    high.append(scores[i])

print(f"Low: {low}\nMedium: {medium}\nHigh: {high}")
```

```python
# Problem 7

ages = [18, 20, 22, 24, 26]
heights = [160, 165, 170, 175, 180]

for i in range(len(ages)):
  print(f"{ages[i]} -> {heights[i]}")

avg_age = sum(ages)/len(ages)
print(f"Average Age - {avg_age}")
avg_height = sum(heights)/len(heights)
print(f"Average Height - {avg_height}")

print("Height appears to increase with age")
```
